Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

Answer-> Scatter plot is the type of plotting which use to repersant the intersection point of x and y axis on the graph

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x,y)
plt.xlabel("this ia the x lable")
plt.ylabel("this ia the y lable")
plt.title("This is a title")
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

Answer-> 
The subplot() function in Matplotlib allows you to plot multiple plots in a single figure, arranged in rows and columns making it convenient for comparing and visualizing multiple datasets or different aspects of a dataset.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create subplots
plt.subplot(2, 2, 1)  # First subplot
plt.plot(x1, y1)
plt.title("Line 1")

plt.subplot(2, 2, 2)  # Second subplot
plt.plot(x2, y2)
plt.title("Line 2")

plt.subplot(2, 2, 3)  # Third subplot
plt.plot(x3, y3)
plt.title("Line 3")

plt.subplot(2, 2, 4)  # Fourth subplot
plt.plot(x4, y4)
plt.title("Line 4")

plt.tight_layout() # Improves the spacing between subplots

plt.show()  # Display the figure
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

Answer-> A bar plot is a type of chart that represents categorical data with rectangular bars, where the length of each bar corresponds to the value it represents

Bar plots are used for various purposes, including:

1 Comparing values

2 Presenting Ranking

```python
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company, profit)
plt.xlabel('Company')
plt.ylabel('Profit (in millions)')
plt.title("A bar plot")
plt.show()
```

```python
plt.barh(company, profit)
plt.xlabel('Profit (in millions)')
plt.ylabel('Company')
plt.title("A horizontal bar plot")
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

Answer-> A box plot summarizes the minimum, first quartile, median, third quartile, and maximum values of a dataset, along with any potential outliers.

Box plots are used for several purposes:

1 Visualizing distribution

2 Visualizing distribution:

```python
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

plt.boxplot(data)
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Value')
plt.title('Box Plot')
plt.show()
```
